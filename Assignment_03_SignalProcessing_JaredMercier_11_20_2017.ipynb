$4$ (f). There are 5 attributes that allow to define the direction of the image with respect to
the patient body: Pixel Spacing, Image Orientation, Image Position, Slice Thickness and
Slice Location. Give the name of the attributes from the DICOM header and describe in
your own words how they describe the geometry of a patient.

Pixel spacing is the distance in the patient between the center of each pixels in both the x and y direction.

Image orientation used is the reference coordinate system (RCS). The 6 parameters used are Right, Left, Anterior, Posterior, Feet, Head. Where they represent the increases and decreases in either the x,y or z direction. Meaning Right defines the direction in which X decreases and Left defines where X increases. 

Image position is the x, y and z direction of the image which is referenced from the top left corner.

Slice thickness is simly how thick the slice will be in mm. For example if a image is taken of an object, and the slice thickness is 0.625mm, the sequence of images will be 0.625mm thick through the object. 

Slice location simply describes the position of the image plane.
