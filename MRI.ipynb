```python
%matplotlib inline
from skimage import novice, data, io, exposure
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import dicom


RefDs = dicom.read_file('IM_0001.dcm')
print(RefDs)
```

```python
print(str(RefDs.EchoTime))
print(str(RefDs.RepetitionTime))
print(str(RefDs.FlipAngle))
```

```python
ConstPixelDims = (int(RefDs.Rows), int(RefDs.Columns))
ArrayDicom = np.zeros(ConstPixelDims, dtype=RefDs.pixel_array.dtype)
ds = dicom.read_file('IM_0001.dcm')
ArrayDicom[:,:]=ds.pixel_array



fig=plt.figure(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.axes().set_aspect('equal') #this makes the image square
plt.set_cmap(plt.gray()) #grayscale
x=np.arange(RefDs.Rows)
y=np.arange(RefDs.Columns)

plt.pcolormesh(x, y, np.flipud(ArrayDicom[:, :]))
plt.axis([x.min(),x.max(),y.min(),y.max()])
print('Scanning sequence: Gradient Recalled '+str(RefDs.ScanningSequence))
```
